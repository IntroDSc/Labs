## Data Wrangling

### Troca dos nomes das colunas

```python
dados.columns = ['Idade','Sexo','BMI','Menores',
                 'Fumante','Região','Valor']
```

```python
dados.head()
```

### Recodificação de variáveis categóricas

```python
map_sexo = {'female':'Feminino','male':'Masculino'}
map_regiao = {'southwest':'Sudoeste','southeast':'Sudeste',
             'northwest':'Noroeste','northeast':'Nordeste'}
map_yesno = {'yes':'Sim','no':'Não'}
```

```python
dados['Sexo'] = dados['Sexo'].map(map_sexo)
dados['Região'] = dados['Região'].map(map_regiao)
dados['Fumante'] = dados['Fumante'].map(map_yesno)
```

```python
dados.head()
```

### Criação de novas colunas com variáveis categóricas

As seguintes colunas serão criadas:
- *Ideal*: indica se o bmi do beneficiário está entre 18.5 e 24.9
- *Nível*: indica se o bmi é abaixo de 18.5, na faixa ideal, ou acima de 24.9
- *Dependentes*: indica se o beneficiário tem pelo menos uma criança como dependente
- *Faixa etária*: faixa etária do beneficiário
   - *Jovem*: indica se o indivíduo tem até 25 anos
   - *Adulto*: indica se o indivíduo tem de 26 a 45 anos
   - *Meia idade*: indica se o indivíduo tem de 46 a 55 anos
   - *Idoso*: indica se o indivíduo tem mais de 55 anos

```python
dados['Ideal'] = dados['BMI'].apply(
    lambda x: 'Sim' if (x >= 18.5 and x <= 24.9) else 'Não')
```

```python
dados['Nível']= pd.cut(dados['BMI'], [0,18.5,24.9,100],
                               labels = ['Baixo','Ideal','Alto'])
```

```python
dados['Dependentes'] = dados['Menores'].apply(
    lambda x:'Sim' if x > 0 else 'Não')
```

```python
dados['Faixa etária'] = pd.cut(dados['Idade'], [0,25,45,55,100],
                               labels = ['Jovem','Adulto',
                                         'Meia idade','Idoso'])
```

```python
dados.head()
```

```python
# Número de linhas final
linhas = len(dados.index)
```

## Análise Exploratória

### Descrição da amostra

#### Variáveis categóricas

As variáveis categóricas incluídas no estudo são:
   - Sexo
   - Fumante
   - Região
   - Ideal (BMI ideal)
   - Dependentes
   - Faixa etária

```python
(pd.crosstab(dados['Sexo'],dados['Fumante'],
             margins=True)/linhas).round(3)*100
```

A tabela mostra que:
- quase 80% dos beneficiários não são fumantes
- o número de homens e mulheres na amostra é praticamente o mesmo
- uma maior proporção de homens é de fumantes do que de mulheres

```python
df = (pd.crosstab(dados['Sexo'],dados['Dependentes'],
                  margins=True)/linhas).round(3)*100

df['% com dependentes por gênero'] = (df['Sim']/(df['Sim']+df['Não'])).\
                                     round(3)*100
df
```

Percebemos também que a maior parte dos beneficiários tem dependentes, e que proporção de homens e mulheres com dependentes é praticamente a mesma.

```python
(pd.crosstab(dados['Faixa etária'],
             [dados['Sexo'],dados['Fumante']],
             margins=True)/linhas).round(3)*100
```

Os resultado apresentados acima indicam que:
- a maior parte dos beneficiários é de adultos (39,8%), sendo praticamente o dobro da proporção de jovens e beneficiários de meia idade (22,9% e 21,2%)
- A distribuição de beneficiários em cada faixa etária não parece variar substancialmente em cada subgrupo (homens ou mulheres fumantes ou não)

```python
(pd.crosstab([dados['Sexo'],dados['Fumante'],dados['Ideal']],
             dados['Região'],
             margins=True)/linhas).round(3)*100
```

```python
(pd.crosstab(dados['Faixa etária'],dados['Região'],
             margins=True)/linhas).round(3)*100
```

Os resultados acima indicam que os beneficiários estão relativamente bem distribuídos por todas as regiões dos EUA.

```python
df = (pd.crosstab(dados['Faixa etária'],
            [dados['Ideal']],
            margins=True)/linhas).round(3)*100

df['BMI ideal na faixa etária'] = (df['Sim']/(df['Sim']+df['Não'])).\
                                  round(3)*100
df
```

A tabela acima indica que:
- 83% dos beneficiários têm BMI fora da faixa ideal
- jovens são os que têm a maior proporção de beneficiários com BMI no nível ideal
- quanto mais alta a faixa etária, menor o percentual de beneficiários com BMI no nível ideal

Segue uma tabela com o percentual de beneficiários por cada subgrupo definido pelas variáveis categóricas da base de dados.

```python
(pd.crosstab([dados['Sexo'],dados['Faixa etária']],
             [dados['Fumante'],dados['Ideal'],dados['Dependentes']],
             margins=True)/linhas).round(3)*100
```

As seguintes conclusões podem ser derivadas da tabela:
- A maior parte dos beneficiários é composta por indivíduos que não são fumantes e que não estão na faixa ideal de BMI (66,7%)
- A faixa etária predominante é a de adultos (39,8%), seguida pelas de jovens e de meia idade, com percentuais muito próximos (22,9% e 21,2%)
- Não parece haver diferenças substanciais entre homens e mulheres

#### Valor do seguro por grupos de beneficiários

A seguir, mostramos os valores pagos mais altos e mais baixos por faixa etária. Não há, aparentemente, valores extremos muito mais altos ou baixos do que os demais em cada faixa.

```python
def extremos(df,var,n,baixos):
    return df.sort_values(by=var, ascending=baixos)[0:n]
```

```python
dados.groupby('Faixa etária').\
               apply(extremos,'Valor',5,False)[['Idade','BMI','Menores','Valor']].round(1)
```

```python
dados.groupby('Faixa etária').\
               apply(extremos,'Valor',5,True)[['Idade','BMI','Menores','Valor']].round(1)
```

Nas próximas tabelas, comparamos os valores médios para diferentes subgrupos da amostra.

```python
dados['Valor'].groupby([dados['Sexo'],dados['Fumante']]).\
               agg([('total','count'),('média','mean'),('desv. pd.','std'),'max','min']).\
               T.round(1)
```

De acordo com a tabela, as diferenças nos valores médios pagos são muito maiores para fumantes em relação a não fumantes, do que para homens em relação a mulheres.

```python
dados['Valor'].groupby(dados['Faixa etária']).\
               agg([('total','count'),('média','mean'),('desv. pd.','std'),'max','min']).\
               T.round(1)
```

Como esperado, o valor médio pago pelo seguro aumenta com a faixa etária do beneficiário.

```python
dados['Valor'].groupby([dados['Faixa etária'],dados['Ideal']]).\
               agg([('total','count'),('média','mean'),('desv. pd.','std'),'max','min']).\
               round(1)
```

Como mostra a tabela acima, independente da faixa etária, o valor médio pago pelo beneficiário é substancialmente menor quando seu BMI está na faixa ideal.

```python
dados['Valor'].groupby([dados['Faixa etária'],dados['Dependentes']]).\
               agg([('total','count'),('média','mean'),('desv. pd.','std'),'max','min']).\
               round(1)
```

Da mesma forma, independente da faixa etária, o valor médio pago pelo beneficiário tende a ser maior quando ele(a) tem dependentes.

#### Descritivas completas

A seguir, são apresentadas as estatísticas descritivas completas por gênero e status de fumante.

```python
dados.groupby([dados['Sexo'],dados['Fumante']]).describe().round(1).T
```

A tabela a seguir reune um subconjunto das estatísticas descritivas apresentadas acima.

```python
por_grupo = dados.groupby(['Sexo','Fumante'])

func_quant = [('n','count'),('média','mean'),('desv. pd.','std'),
              ('mediana','median')]

por_grupo.agg({'Idade':func_quant,
               'BMI':func_quant,
               'Menores':func_quant,
               'Valor':func_quant}).round(1).T
```

Por fim, na tabela abaixo, apresentamos as médias de cada variável quantitativa por gênero, status de fumante, e nível de BMI (ideal ou não).

```python
dados.pivot_table(index=['Sexo','Fumante'],
                  columns='Ideal',
                  margins=True).round(1).T
```

Pode-se ver, como já mencionado antes, que:
- há poucas diferenças nas médias de BMI, Idade e número de dependentes em função de gênero e status de fumante
- em média, os valores pagos diferem pouco em função do gênero do beneficiário
- fumantes pagam em média por seus seguros mais do que o dobro do que pagam não fuamantes
- beneficiários com nível ideal de BMI pagam em média menos do que os com níveis fora da faixa ideal
- essas conclusões valem para todas as faixas etárias, assim como para o grupo de beneficiários como um todo

#### Conclusão

De acordo com os resultados obtidos, as variáveis com maior impacto (aparente) no valor pago pelo seguro são:
- status de fumante
- nível de BMI (ideal ou não)
- faixa etária
- dependentes

É importante destacar que, nas análises até agora realizadas, apenas foram considerados valores médios. As relações entre variáveis poderão ser melhor exploradas por meio da visualização dos dados em gráficos, que é tópico das próximas seções.

### Visualização dos dados



#### Configuração do matplotlib

```python
plt.rc('figure', figsize=(10, 10))

fontes = {'family': 'sans-serif',
          'style': 'normal',
          'weight': 'medium',
          'size': 12.0}
plt.rc('font', **fontes)
```

#### Distribuição das variáveis quantitativas

```python
fig = plt.figure()

ax1 = fig.add_subplot(221,
                      title='Idade',
                      ylabel='Frequência')
ax2 = fig.add_subplot(222,
                      title='BMI',
                      ylabel='Frequência')
ax3 = fig.add_subplot(223,
                      title='Número de Dependentes',
                      ylabel='Frequência')
ax4 = fig.add_subplot(224,
                      title='Valor Pago',
                      ylabel='Frequência')

plt.subplots_adjust(wspace=.3, hspace=.3)

dados['Idade'].plot.hist(ax=ax1,
                         bins=15,
                         color='lightsteelblue' )

dados['BMI'].plot.hist(ax=ax2,
                         bins=10,
                         color='mediumslateblue' )

dados['Menores'].value_counts().\
                 plot.bar(ax=ax3,
                          color='slateblue',
                          rot=0)

dados['Valor'].plot.hist(ax=ax4,
                         bins=30,
                         color='darkslateblue' )
```

O histograma de <u>idade</u> mostra que:
- o número de beneficiários com até 20 anos de idade é praticamente o dobro do que os de beneficiários nas faixas de idade superiores
- o número de beneficiários é razoavelmente uniforme para as faixas superiores a 20 anos

No segundo histograma, verificamos que os valores de <u>BMI</u> parecem estar normalmente distribuídos em torno de sua média (30,7).

O gráfico de <u>número de dependentes</u> indica que a grande maioria dos beneficiários não têm dependentes menores de idade. Como esperado, a frequência do número de dependentes é decrescente, havendo poucos indivíduos com mais de 3 dependentes.

O histograma do <u>valor pago pelo seguro</u> é o mais interessante. De acordo com o gráfico:
- poucos beneficiários pagam valores acima de 50.000,00
- a grande maioria paga valores abaixo de 15.000,00
- a frequência de beneficiários parece ser razoavelmente uniforme para valores entre 15.000,00 e 50.000,00.

O histograma a seguir, para a faixa de valores acima de 15.000,00, mostra que o número de beneficiários com valores acima de 50.000,00 é realmente diminuto. Por esses indivíduos serem considerados atípicos, serão removidos das análises posteriores.

```python
dados[dados['Valor'] > 15000]['Valor'].plot.hist(bins=30,
                                                 title='Histograma para Valores acima de $15 mil')
plt.ylabel('Frequência')
```

        </ul></li></br>
    <li>BMI:
        <ul>
            <li>BMI também parece estar relacionado com o valor do seguro, embora de maneira menos clara.</li>

            <li>De fato, no diagrama de dispersão, fica claro que os valores pagos por beneficiários com BMI ideal variam numa faixa com valor máximo mais baixo do que os valores pagos por indivíduos no outro grupo.</li>
        </ul></li></br>
    <li>Dependentes:
        <ul>

        </ul></li></br>

#### Fatores com influência no valor do seguro

Nesta seção, exploramos em maior detalhe os fatores quantitativos e qualitativos que parecem afetar o valor do seguro pago pelo beneficiário.

Abaixo, mostramos as relações distintas entre BMI e valor para fumantes e não fumantes. Como dito antes, para fumantes, quanto maior o BMI, maior o valor pago. Isso parece valer mesmo quando o nível de BMI está na faixa ideal.

```python
g = sns.FacetGrid(dados, col='Fumante', hue='Nível')
g = g.map(plt.scatter,'BMI', 'Valor', alpha=.2).add_legend()
```

Nos diagramas abaixo, utilizamos o status de fumante para definir a cor dos pontos. Observamos que a relação entre idade e valor muda consideravelmente com o nível de BMI do beneficiário (baixo, ideal, alto). Para niveis baixos e ideal de BMI, parece haver duas linhas de tendência associando o valor à idade: uma mais baixa (em vermelho), para não fumantes; e uma mais alta (em azul), para fumantes. No entanto, há uma terceira linha de tendência, acima das outras duas, quando o nível de BMI do beneficiário é alto. Investigaremos a relação para este grupo de indivíduos mais detalhadamente nos próximos gráficos.

```python
g = sns.FacetGrid(dados, col='Nível',hue='Fumante')
g = g.map(plt.scatter,'Idade', 'Valor', alpha=.2).add_legend()
```

Nos próximos diagramas, apenas beneficiários na faixa de BMI alto são considerados. 

```python
g = sns.FacetGrid(dados[dados['Nível']== 'Alto'], col='Dependentes',hue='Fumante')
g = g.map(plt.scatter,'Idade', 'Valor', alpha=.2).add_legend()
```

```python
g = sns.FacetGrid(dados, col='Sexo',hue='Fumante')
g = g.map(plt.scatter,'Idade', 'Valor', alpha=.2).add_legend()
```

```python
g = sns.FacetGrid(dados[dados['Nível']== 'Alto'], col='Região',hue='Fumante')
g = g.map(plt.scatter,'Idade', 'Valor', alpha=.2).add_legend()
```

De acordo com os gráficos, gênero, ter ou não dependentes menores de idade, e região de residência não influenciam a relação entre idade e valor pago pelo seguro. As três linhas de tendência se mantiveram praticamente inalteradas em todos os diagramas.

## Conclusão


De acordo com as análises apresentadas acima, o valor pago pelo beneficiário é função principalmente de sua idade e do seu nível de BMI. No entanto, as relações entre esses fatores e o fato de o indivíduo fumar ou não mudam consideravelmente. De forma geral, fumantes tendem a pagar, em média, valores bem mais altor por seus seguros. A diferença média de valor entre fumantes e não fumantes não parece variar em função da idade.

Observamos também a presença de três faixas distintas relacionando idade ao valor pago principalmente para indivíduos com BMI alto. Nossas análises não identificaram a causa para isso, que provavelmente está associada a diferenças numa ou mais variáveis categóricas. É provável que essas variáveis não tenham sido incluídas na base de dados, devendo ser objeto de estudos futuros.
