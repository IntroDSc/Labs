Nosso objetivo é aprender como tratar váriáveis que são cadeias de caracteres em um *DataFrame*. Em muitos casos, essas variáveis representam categorias (ou grupos) a que a entidade de interesse (pessoa, carro, empresa, etc.) pertencem.

Quando necessário, você pode consultar as seguintes seções da documentação do `pandas`:

 - [Working with text data](https://pandas.pydata.org/docs/user_guide/text.html)
 - [Categorical data](https://pandas.pydata.org/docs/user_guide/categorical.html)

A base de dados com que vamos trabalhar é a de passageiros do Titanic. Ela está disponível na seguinte página: (https://gist.githubusercontent.com/fyyying/4aa5b471860321d7b47fd881898162b7/raw/6907bb3a38bfbb6fccf3a8b1edfb90e39714d14f/titanic_dataset.csv)


```python
import numpy as np
import pandas as pd
pd.set_option('display.float_format', '{:.3f}'.format)
```

**1.** Leia a base de dados para um *DataFrame*. 

```python
dados = pd.read_csv("https://gist.githubusercontent.com/fyyying/4aa5b471860321d7b47fd881898162b7/raw/6907bb3a38bfbb6fccf3a8b1edfb90e39714d14f/titanic_dataset.csv")
dados.head()
```

**2.** Gere as informações básicas sobre o *DataFrame* criado. Por exemplo, quantas linhas e colunas ele tem? Que tipos de variáveis estão contidas nas colunas? Há dados faltando? Em que colunas? Quantos? 

```python
dados.info()
```

```python
dados.describe()
```

```python
pd.DataFrame(dados.isna().sum(), columns = ["Valores omitidos"])
```

**3.** Traduza o nome das colunas para o português.

```python
dados.rename(columns = {"PassengerId": "ID", 
                        "Survived": "Sobrevivente", 
                        "Pclass": "Classe", 
                        "Name": "Nome", 
                        "Sex": "Sexo", 
                        "Age": "Idade", 
                        "SibSp": "Irmãos e Esposos", 
                        "Parch": "Pais e Filhos", 
                        "Ticket": "Bilhete", 
                        "Fare": "Tarifa", 
                        "Cabin": "Cabine", 
                        "Embarked": "Local de Embarque"},
            inplace = True)
dados.head()
```

**4**. Crie colunas que correspondam ao primeiro nome da pessoa e a seus sobrenomes. Note que há pessoas listadas como esposas de um outro passageiro. As novas colunas devem conter os nomes das esposas.

```python
strings = dados.select_dtypes(include = ["object"]).columns
dados.loc[:, strings] = dados.loc[:, strings].astype("string")
dados.info()
```

```python
dados["Nome"]
```

```python
nomes = dados["Nome"].str.split(",", expand = True)
nomes.columns = ["Sobrenome", "Nome"]
nomes.head()
```

```python
nomes["Nome"] = nomes["Nome"].str.replace("[DMR][evirs.]*\.* ", "", regex = True)
nomes.head()
```

```python
nomes["Nome"] = nomes["Nome"] + " " + nomes["Sobrenome"]
nomes.drop(columns = "Sobrenome", inplace = True)
nomes.head()
```

```python
nomes["Parênteses"] = nomes["Nome"].str.extract("\(([a-zA-Z \"]*)\)")
nomes
```

```python
nomes["N"] = nomes["Parênteses"].apply(lambda x: len(x.split()) if pd.notna(x) else np.nan)
nomes.sort_values("N")
```

```python
nomes.loc[(nomes["Parênteses"].notna()) & (nomes["N"] > 1), "Nome"] = nomes["Parênteses"]
# = nomes.loc[(nomes["Parênteses"].notna()) & (nomes["N"] > 1), "Parênteses"]
nomes
```

```python
nomes = nomes["Nome"].str.rsplit(n = 1, expand = True)
nomes.columns = ["Nome", "Sobrenome"]
nomes = nomes.astype("string")
nomes
```

```python
dados.drop(columns = "Nome", inplace = True)
dados = pd.concat([dados, nomes], axis = 1)
dados = dados.reindex(columns=['ID', 'Nome', 'Sobrenome','Sobrevivente', 'Classe', 'Sexo', 'Idade', 
                               'Irmãos e Esposos', 'Pais e Filhos', 'Bilhete', 'Tarifa', 'Cabine', 
                               'Local de Embarque'])
dados
```

**5.** Transforme *Sexo* (*Sex*), *Local de Embarque* (*Embarked*) e *Classe*(*Pclass*) em variáveis categóricas, identificando por extenso as categorias correspondentes.

```python
dados["Sexo"] = dados["Sexo"].map({"male": "masculino",
                                   "female": "feminino"},
                                  na_action = "ignore")

dados["Local de Embarque"] = dados["Local de Embarque"].map({"C": "Cherbourg", 
                                                             "Q": "Queenstown", 
                                                             "S": "Southampton"},
                                                            na_action = "ignore")
dados.head()
```

```python
from pandas.api.types import CategoricalDtype

cat_sexo = CategoricalDtype(["feminino", "masculino"], ordered = False)
cat_embarque = CategoricalDtype(["Cherbourg", "Queenstown", "Southampton"], ordered = False)
cat_classe = CategoricalDtype([1, 2, 3], ordered = True)

dados["Sexo"] = dados["Sexo"].astype(cat_sexo)
dados["Local de Embarque"] = dados["Local de Embarque"].astype(cat_embarque)
dados["Classe"] = dados["Classe"].astype(cat_classe)
```

```python
dados.info()
```

```python
dados[["Sexo", "Local de Embarque", "Classe"]].describe()
```

```python
pd.pivot_table(dados,
               values = "Sobrevivente",
               index = ["Sexo", "Local de Embarque"], 
               columns = ["Classe"],
               aggfunc = ["count"],
               margins = True,
               dropna = True)
```

```python
dados["Sexo"].value_counts()
```

```python
dados["Local de Embarque"].value_counts()
```

```python
dados["Classe"].value_counts()
```

**6.** Calcule as estatísticas descritivas básicas para todas as variáveis numéricas do *DataFrame*.

```python
dados.select_dtypes(include = ["float", "int"]).columns
```

```python
dados[['Sobrevivente', 'Idade', 'Irmãos e Esposos', 'Pais e Filhos', 'Tarifa']] \
    .agg(
        {'Sobrevivente': ["sum", "mean"],
         'Idade': ["min", "max", "median", "mean", "std", "skew"], 
         'Irmãos e Esposos': ["min", "max", "median"],
         'Pais e Filhos': ["min", "max", "median"], 
         'Tarifa': ["min", "max", "median", "mean", "std", "skew"]})
```
