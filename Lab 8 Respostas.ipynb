Neste laboratório, continuaremos nossa prática de obtenção, limpeza e tratamento de dados. Nosso objetivo é gerar um *DataFrame* com o IDH de todos os países e seus PIBs, conforme listado nas seguintes páginas da Wikipedia:

 - [Lista de países por Índice de Desenvolvimento Humano](https://pt.wikipedia.org/wiki/Lista_de_pa%C3%ADses_por_%C3%8Dndice_de_Desenvolvimento_Humano)
 - [Lista de países por PIB nominal](https://pt.wikipedia.org/wiki/Lista_de_pa%C3%ADses_por_PIB_nominal)

O *DataFrame* deverá conter o nome do país, o continente a que pertence, o valor do IDH, e o valor nominal do PIB.

**Dados de IDH**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
pd.set_option('display.float_format', '{:,.3f}'.format)
```

```python
url = "https://pt.wikipedia.org/wiki/Lista_de_pa%C3%ADses_por_%C3%8Dndice_de_Desenvolvimento_Humano"
html = pd.read_html(url, attrs = {'class': 'multicol'})
```

```python
idh = pd.DataFrame(columns = ["Ranking", "País", "IDH"])

for i in range(0, 4): 
    table = pd.DataFrame(html[i])
    table.columns = ["Ranking", "Ranking2", "País", "IDH", "IDH2"]
    table.drop(columns = ["Ranking2", "IDH2"], inplace = True)
    idh = idh.append(table.loc[table["Ranking"].str.fullmatch("[0-9]+"), :])
    
idh.reset_index(drop = True, inplace = True)
idh
```

```python
idh.info()
```

```python
idh = idh.astype(dtype = {"Ranking": "int", "País": "string", "IDH": "float"})
idh.info()
```

```python
idh["IDH"] = idh["IDH"]/1000
idh
```

**Dados de PIB de acordo com a lista do Banco Mundial (2018)**

```python
url = "https://pt.wikipedia.org/wiki/Lista_de_pa%C3%ADses_por_PIB_nominal"
html = pd.read_html(url, match = "País", flavor = "bs4", thousands = "\xa0", decimal = ",", encoding = "utf-8")
```

```python
pib = pd.DataFrame(html[2])
pib.drop(index = 0, inplace = True)
pib
```

```python
pib.loc[pib["Posição"].str.contains("\D"), :]
```

```python
pib.loc[pib["Posição"].str.contains("\D"), "Posição"] = np.nan
```

```python
pib.loc[pib["País"].isin(["Hong Kong", "Macau", "Palestina", "Kosovo"]), :]
```

```python
pib.loc[pib["País"].str.contains("\d"), :]
```

```python
pib.loc[pib["País"].str.contains("\d"), "País"] = \
    pib["País"].str.extract("([\w ]*)\[.*", expand = False)
```

```python
pib.loc[pib["País"].isin(["Marrocos", "Tanzânia", "Chipre", "Geórgia", "Moldova"]), :]
```

```python
pib.loc[pib["PIB (Millhões de US$)"].str.contains("\D"), :]
```

```python
pib.loc[pib["PIB (Millhões de US$)"].str.contains(" "), :]
```

```python
pib.loc[pib["PIB (Millhões de US$)"].str.contains(" "), "PIB (Millhões de US$)"].to_list()
```

```python
pib.loc[pib["PIB (Millhões de US$)"].str.contains(" "), "PIB (Millhões de US$)"] = \
    pib["PIB (Millhões de US$)"] \
    .str.extract("([\d\xa0,]+) [.]*", expand = False) \
    .str.replace("\xa0", "") \
    .str.replace(",", ".")
```

```python
pib.loc[pib["Posição"].isin(["27", "130", "150", "165"]), :]
```

```python
pib = pib.astype(dtype = {"Posição": "float", "País": "string", "PIB (Millhões de US$)": "float"})
pib.info()
```

**1.** Que países têm os 10 maiores e menores valores de IDH e PIB? Parece haver uma relação entre as duas variáveis? 

```python
pib_10 = df.sort_values("PIB (Millhões de US$)").iloc[:10]
pib_10 = pd.concat([pib_10, df.sort_values("PIB (Millhões de US$)", ascending = False).iloc[:10]])
pib_10.sort_values("PIB (Millhões de US$)", ascending = False, inplace = True)
pib_10
```

**2.** Calcule o IHD e o PIB médios por continente, assim como seus valores mínimos e máximos, e os seu desvios-padrão.

```python
df \
    .loc[:, ["Continente", "IDH", "PIB (Millhões de US$)"]] \
    .groupby(by = ["Continente"], 
             dropna = True) \
    .agg(['mean', 'std', 'min', 'max']) \
    .transpose()
```

**3.** Crie uma coluna que indique se o país tem um PIB muito alto, alto, mediano, baixo, ou muito baixo. Para isso, calcule os percentis de 20% da variável PIB e use os limites de cada faixa para agrupar os países em categorias (*binning*). Caso necessário, consulte a página [Binning Data with Pandas qcut and cut](https://pbpython.com/pandas-qcut-cut.html).

```python
df["Faixa de PIB"] = pd.qcut(df["PIB (Millhões de US$)"], 
                             q = 5, 
                             labels = ["Muito baixo", "Baixo", "Mediano", "Alto", "Muito alto"])
df
```

**4**. Faça o mesmo para a coluna IDH.

```python
df["Faixa de IDH"] = pd.qcut(df["IDH"], 
                             q = 5, 
                             labels = ["Muito baixo", "Baixo", "Mediano", "Alto", "Muito alto"])
df
```

**5.** Calcule que proporção dos países em cada faixa de PIB estão na faixa correspondente de IDH.

```python
tabela = pd.pivot_table(df, 
                        values = "País",
                        index = ["Faixa de PIB"], 
                        columns = ["Faixa de IDH"],
                        aggfunc = ["count"],
                        dropna = True,
                        margins = True,
                        margins_name = "Todos")
tabela
```

```python
multi = tabela.columns
multi
```

```python
pd.set_option('display.float_format', '{:,.1f}'.format)

multi = multi.set_levels(["Percentual em cada faixa de IDH"], level = 0)
tabela.columns = multi

tabela = tabela.apply(lambda x: x * 100 / tabela.loc[:, pd.IndexSlice["Percentual em cada faixa de IDH", "Todos"]])
tabela
```

```python
sns.heatmap(tabela.loc["Muito baixo":"Muito alto", 
                       pd.IndexSlice[:, "Muito baixo":"Muito alto"]] \
                  .transpose() \
                  .droplevel(level = 0) \
                  .reindex(labels = ["Muito alto", "Alto", "Mediano", "Baixo", "Muito baixo"]),
            cmap = "viridis",
            annot = True)

plt.title("Heatmap para proporções de países\nem faixas de IDH por faixa de PIB", y = 1.05, fontsize = 12)
plt.yticks(rotation=0);
```
